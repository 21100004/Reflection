# Reflection from a flat surface

Reflection is well known since millenia. We do expect QED to explain this simple pnenomena. Consider a thin beam of light falling on a reflecting mirror. The beam is reflected though some of it also passes straight through. For now, we consider only the reflected part and we'll talk about the transmission later.

In school we learnt about the law of reflection: the angle of reflection must be equal to the angle of incidence. The angles are defined with respect to the surface normal. Stated simply


\begin{equation}
\theta_{2}=\theta_{1}
\label{eq:reflection}
\end{equation}

where $\theta_1$ and $\theta_2$ define the angles of incidence and reflection respectively.

For a beam of light, this geometrical picture is so familiar to us that it's taken for granted. But things start getting more interested, at least from the physicist's perspective, when we attenuate the light beam by placing for exaample a medium in between that has a high optical density blocking most of the light. The attenuator could be so effective that perhaps a single photons emerges from it's other end. This was a single photons impinges on the flat mirror surface. The next photon would emerge at some later time so that if you were to take a snapshot of the region in the mirror's vicinity, you would "see" only a single photon.
If light comprises a singe photon, the situation becomes more interesting. By interesting I mean counter-intuitive for the person who has not undergone a formal physics training at the graduate level! How does a photon know which path to take after reflection? This is the key question, so let me state it again.

<font color=red> How does the photon know which path to take upon reflection? </font>

In order to understand this question and attempting to answer it, let's propose a suitable geometry of the problem. A flat mirrow is placed at $y=0$ and is along the $x$ axis. It's length is taken to be $10$ units. Our light cource is placed at location $A$ and to avoid complications, we assume it is a genuine single photon source produced from some highly sophisticated technology that we admire. We place a detector at another location on the same side of the mirror. The detector fires and "sees" the photon at a later time. The detection location is marked as $B$. The detector can be a photomultiplier tube or an avalanche photodetector that is capable of also detecting single counts of photons. If we were waiting to detect a photon at $B$ what path must the single photon take as it emerges from $A$, falls onto the mirror and is reflected onward to $B$?

We specify the locations $A$ and $B$ and the length of the mirror. We also break up the mirror into $10$ equal parts. This sectioning of the mirror will be useful later.

```python
sourceX = 0
sourceY = 10
photoX = 10
photoY = 10
mirrorLen = 10
numPiece = 10
fig = reflection.placeSAndP(mirrorLen, numPiece, sourceX, sourceY, photoX, photoY)
```

The sections are labeled in a specific fashion. The total length is $10$ m. $x_0$ shows the section starting at $x=0$ and ending at $x=1$ and so on. Therefore $x_n$ is the section starting at length $x=n$ and ending at $x=n+1$. For mirrors of different lengths and different number of total sections, the labeling will change accordingly. We will also use the starting index as the overall section label. You can try this out for yourself. Point $A$ is a light source shown as a red star and $B$, the photodetector as a green diamond. The actual path taken by a photon, conforming to the law of reflection mentioned above, is shown below.

```python
fig = reflection.classicalPath(mirrorLen, numPiece, sourceX, sourceY, photoX, photoY)
```

How does the single photon know that it must impinge on the section labeled $x_3$ in order to reach $B$? <font color=red>In fact the photon doesn't know which path to take. It is dumb though not an absolute idiot.</red> This is described below.

QED and other quantum field theories are sometimes described by a theory called *path integral formalism*. We can view the reflection of light, whose classical exposition is presented above, in light of path integrals. In this formalism, the photon is not constrained to take up only the classical path shown in red in the figure below. It is rather unintelligent. Remarkably wise in another way! The single photon can take up one of an infinite number of paths. The realm of possibilities is infinite. And it does so without splitting at all. The photon is indivisible.

Shown below are only some of the possible paths. Each path is sometimes called a *history*. There are infinite histories possible and we cannot show all of them. We choose only $10$ which is the number of sections we've imagined the mirror to be broken in. Shown in black is the history corresponding to reflection from (the exact middle of) one of the sections, $x_1$ through $x_{9}$. The history that is actually realized in reality is shown in <font color=red>red</font> which bounces off from some point in $x_3$.

```python
reflection.QEDPath(mirrorLen, numPiece, sourceX, sourceY, photoX, photoY)
```

So far we have done nothing clever. After all, we can draw as many paths (histories) as we like but the *real* path is the <font color=red>red</font> one. So why are the multiple histories plausible? They don't mean anything. In fact QED's path integral formalism let's us pick one path from the multiplicity of paths. Each path is assigned a particular weight (which we'll epxlain more quantitatively later).

## Least time

One possible cause of the photon choosing the <font color=red>red</font> path over all other possibilities may be that it takes the least time. Could the photon be in a hurry reaching for $B$. While we don;t know why this should be true, we nevertheless analyze this situation graphically. Light has a finite speed. In vacuum the speed is denoted by $c$ and equal to $3\times 10^8$ meters per second. The path lengths for the ten paths shown in the Figure are all different. The time for each photon path is simple: the path length divied by $c$. Here is what the calculation results in.

```python
speed = 3 * 1e8
```

```python
times = reflection.createTimes(mirrorLen, numPiece, sourceX, sourceY, photoX, photoY, speed)
reflection.plotTimes(mirrorLen, numPiece, times)
```

We observe that indeed, as we predicted the classical path corresponding to bouncing off of the photon from the section $x_3$ has the least time. Classically light follows this *least time* path and so would a single photon. Hence we conclude that:

<font color=red>A photon takes the path in a medium for which the time taken to traverse the path is minimum</font>, though we have no justification of why this could be true. The path integral formalism of QED describes the paths in a more unifying and mentally coherent picture (though it still does not explain the *why*).

The time curve shown above gets smoother and more realistic as we change the value of the variable numPiece. Go back and try this for yourself.

## Path integral formalism

Before we introduce the path integral picture, we like to introduce the concept of a phasor. I assume you are familiar with complex numbers. A phasor is merely a complex number which is drawn as an arrow. The length of the arrow is the *amplitude*. The arrowhead also shows that there is more to a phasor, it also has a direction. The direction in two dimensions is specified by an angle the phasor makes with some fixed axis. We arbitrarly choose this reference axis to lie horizontally. The angle is also called the *phase*. We show below how a phasor is drawn. Adjust the amplitude and the phase to see how the picture changes.

```python
amp = 2
phase = 60
reflection.phasor(amp, phase)
```

What does the path have to do with phasors? We have noted that each path takes a particular time which is calculated in the case above by purely geometric means. Imagine that each path carries with it a tiny clock which ticks as the photon emanates from $A$ and stops when it reaches $B$. Each clock shows what is called the *proper time*. All clocks are synchronized, are ideal and we don't get into philosophical questions such as who carries the clock, the photon or the underlying spapcetime. 

As the photon traverses a path, the phasor advances in time. Its amplitude remains fixed while its phase changes. The phase is proportional to the time passed since the photon had left the source. The proportionality between the phase and the time is determined by a frequency which tells us how fast the arrow of the phasor rotates. This frequency is interestingly linked with the color (or wavelength of light). For the time being we assume that we are dealing with photons of just one color, say green.

\begin{equation}
\text{phase} = \text{frequency}\times\text{proper time}.
\end{equation}

Suppose we consider a particular path, say $x_0$ in our reflection from a flat mirror. The time it takes for a photon adopting this path in going from $A$ to $B$ is approximately $69$ ns. The diagram below shows how the phasor rotates as the clock of path $x_0$ ticks. We have set the frequency of the phasor arbitrarily at some value which is not important for the time being. Suppose the phasor makes one complete rotation of $360$ degrees in {{numPiece}} ns. Hence the frequency of rotation is $36$ degrees per nanosecond.

```python
amp = 2
freq = 2
fig, updatePhasor, fargs, frames, interval, blit = reflection.timePhasor(amp, freq)
from matplotlib import animation
from IPython.display import HTML
%matplotlib inline
anim = animation.FuncAnimation(fig, updatePhasor, fargs = fargs, frames = frames, interval = interval, blit = blit)
HTML(anim.to_jshtml())
```
